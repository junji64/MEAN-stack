## MongoDB : RDB vs. NoSQL

**관계형 데이터베이스(RDB, Relational DB)**는 데이터를 Column(열)과 Row(행)로서 나타내었다. 각 Column은 데이터의 타입을 표현하고, 각 Row는 하나의 레코드 Entry(Record)를 의미한다. 

반면, **NoSQL(Not Only SQL)** 데이터베이스 기술은 관계형 데이터베이스에서 사용하는 열과 행 대신 JSON 문서에 정보를 저장하는 데이터베이스 유형이다. 결과적으로 NoSQL 데이터베이스는 유연하고 확장 가능하며 현대 기업의 데이터 관리 요구에 신속하게 대응할 수 있도록 구축되었습니다.
**MongoDB**는 NoSQL의 하나인 **문서형 데이터베이스**의 대표적인 케이스로, RDB와는 다른 구조적 특징을 가지고 있다.

- Row는 있지만 Column이 없다.
 - 정확히는 Row 속에 RDB에서의 Column 역할이 포함되어 있다고 볼 수 있다.
 - 문서형 데이터베이스에서 하나의 Row는 데이터 타입과 각 데이터 타입에 부합하는 데이터를 모두 소유한다.
 - 이전 RDB에서는 모든 Row가 동일한 Column에 종속적이었던 것과 달리, 문서형 데이터베이스는 각 Row가 서로 다른 Column의 종류, 개수를 소유하는 것이 가능하다.
 
 
- 무엇이 좋은 점인가?
 - 엄격한 구조적 지배에서 벗어나서 확장성이 매우 높다고 볼 수 있다.
 - 기존 RDB는 Column이 다른 구조화된 자료구조를 가지는 것이 불가능 했으며, 이를 외래키(Foreign Key)로 해결하고자 하였다. 하지만 문서형 데이터베이스는 어떠한 데이터 타입이라도 수용 가능한 **JSON** 이라는 형태의 구조를 가진다.
 
 
- 관계형 데이타베이스 => 문서형 데이타베이스
    - 테이블 => **컬렉션(collection)** 
    - 레코드 => **문서객체(document)**

### JSON 문법

* 데이터는 이름:값 쌍으로 넣습니다.    

ex)  { “name”:“Jun” }


* 데이터는 쉼표로 구분됩니다.            

ex)  { “name”:“Jun”, “age”:55 }


* 객체(object)는 중괄호 { } 에 넣습니다. 

ex) { "employee": { "name":"John", "age":30, "city":"New York" } }


* 배열(array)은 대괄호 [ ] 에 넣습니다. 

ex)  { "employees": [ "John", "Anna", "Peter" ] }


* 값으로는 문자열, 숫자, JSON 객체, 배열, 불리안(boolean), null 이 올 수 있다. 

ex) { “item”: “shirt”, “sale”: true, "brand":null }

## CRUD using Mongoose  ([ref](https://www.youtube.com/watch?v=swWRUvluSkE&list=PLGquJ_T_JBMQ1C0Pp41sykceli8G1UGtg))

Mongoose는 mongodb와 쉽게 상호 작용할 수 있게 해주는 노드 모듈입니다. 
스키마(Schema)를 생성하고 데이터를 조작하기 위한 API 형태의 추상화를 제공합니다. 
스키마는 mongodb 에서 사용할 수 있는 다양한 데이터 유형에 대한 키 및 값 쌍을 생성합니다. (for detail see http://mongoosejs.com )

````javascript
const BookSchema = new Schema({
    title: String,
    published: {
        type: Date,
        default: Date.now
    }
    keywords: Array,
    published: Boolean
});
````

#### What does Mongoose do?
Create Object Reference : Mongoose는 mongoDB와 상호 작용할 때 사용하기 쉬운 객체 참조를 생성합니다. 데이터베이스가 코드 내에서 모델링된다는 것입니다. 

#### How to use Mongoose?
제일 먼저 mongoose 패키지를 설치합니다. (예: >> npm install mongoose )
  1. mongoose 모듈을 require()로 가져옵니다.
  2. mongoose.schema 를 이용하여 스키마를 생성합니다. (예: mongoose.Schema( ) )
  3. schema 와 다른 파일을 사용하려면 schema 의 인스턴스를 만들어야 합니다. mongoose 를 사용할 때 우리는 이 인스턴스를 모델이라고 부릅니다. (예: mongoose.model( ))
  4. 이제 node 에 주어진 편리한 export 방법을 사용하여 module 을 사용할 수 있게 됩니다. 
  5. 이제 간단히 이 스키마를 require( ) 로 가져와서  새 스키마 및 기타 파일을 사용할 수 있습니다. 

````javascript
//   /myapp/models/Book.model.js
//
const mongoose = require('mongoose');        // 1
const BookSchema = new mongoose.Schema({     // 2
    title: String,
    keywords: Array,
    published: Boolean
});
module.exports = mongoose.model('Book', BookSchema);   // 3,4
````

- Now…. Edit  routes/index.js as

````javascript
var express = require('express');
var router = express.Router();
var Book = require('./book.js');
var mongoose = require('mongoose');

mongoose.connect('mongodb://localhost:27017/bookstore');

/* GET home page. */
router.get('/', function(req, res) {
  res.render('index', { items: [] });
});

router.get('/get-data', function(req, res) {
  Book.find( function(err, docs) {
    if(err) {
      res.status(500).send({ error: 'database failure' });
      return;
    }
    res.render('index', {items: docs})
  });
});


router.post('/insert', function(req, res) {
  var book = new Book();
  book.title = req.body.title;
  book.content = req.body.content;
  book.author = req.body.author;

  book.save(function(err){
    if(err) {
      res.status(500).send({ error: 'database failure' });
      return;
    }
    res.redirect('/');
  });
});

router.post('/update', function(req, res) {
  Book.updateOne({_id: req.body.id},
    {title:req.body.title, 
    content:req.body.content, 
    author: req.body.author},
    function(err) {
      if(err) {
        res.status(500).send({ error: 'database failure' });
      };
      res.redirect('/');
  });
});


router.post('/delete', function(req, res) {
  Book.remove({_id: req.body.id}, function(err, output) {
    if(err) res.status(500).json({ error: 'database failure' });
    console.log(output);
    res.redirect('/');
  });
});

module.exports = router;
````
